```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('respiradores.csv' , sep = ",")

x = df.MES
y = df.TOTAL
plt.bar(x,y, 1.5, align = "center", edgecolor = "black")
plt.title("COMPRA DE RESPIRADORES POR MES")
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.show()
```
