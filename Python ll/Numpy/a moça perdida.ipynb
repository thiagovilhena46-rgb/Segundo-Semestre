```python
titanic = pd.read_csv("titanic.csv", sep=",")
def faixa_etaria(linha):
    if linha["Age"] < 12:
        return "Criança"
    elif linha["Age"]< 18:
        return "Adolecente"
    elif linha["Age"]< 65:
        return "Adulto"
    elif linha["Age"] >= 65:
        return "Idoso"
    else:
        return "null"
faixa = titanic.apply(faixa_etaria, axis=1)

titanic["faixa etaria"] = faixa
print(titanic[["Age","faixa etaria"]].head(50))

        
```

```python
def calculate_percentage(val, total):
#Calculates the percentage of a value over a total
    percent = float(val / total)
    beautiful_percent = ("%.2f" % (percent * 100) + "%")
    return beautiful_percent
```
