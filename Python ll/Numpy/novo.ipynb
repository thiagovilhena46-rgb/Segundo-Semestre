```python
import pandas as pd
titanic = pd.read_csv("new_titanic.csv")
```

```python
agrupado = titanic.groupby("Sex")["Age"].mean().to_frame(name = "Media Idade")
   
vivos = titanic[titanic["Survived"] == 1 ].groupby("Sex").count() 
mortos = titanic[titanic["Survived"] == 0 ].groupby("Sex").count() 
agrupado["vivos"] = vivos["PassengerId"]
agrupado["mortos"] = mortos["PassengerId"]
print(agrupado)
```

```python
train = pd.read_csv("train.csv")
mulher = train.Sex[(titanic["Sex"] == "female")].count()
homen = train.Sex[(titanic["Sex"] == "male")].count() 
total = train.Sex[(titanic["Survived"])].count() 
print(round((mulher/total) * 100,2))
print(round((homen/total) * 100 , 2))
```
