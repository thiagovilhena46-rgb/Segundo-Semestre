```python
import numpy as np

n = 4

m = np.random.randint(0,11,(n,n)) # Changed to create an n x n 2D array
print(m,'\n')

for i in range(n):
    for j in range(n):

        v = 0

        if i+1 < n:
            v += m[i+1,j]

        if i-1 >= 0:
            v += m[i-1,j]

        if j+1 < n:
            v += m[i,j+1]

        if j-1 >= 0:
            v += m[i,j-1]

        print(v,end=" ")
    print()
```
